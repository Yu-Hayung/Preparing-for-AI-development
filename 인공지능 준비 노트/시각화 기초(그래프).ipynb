# Graph Visualization

---
## Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline  # 매직 함수 (쥬피터에만 사용가능한 함수 첫 패키지 풀러오기)
```

---
## Basic Attributes

alpha : 투명도   
kind : 그래프 종류 'line', 'bar', 'barh', 'kde'  
logy : Y축에 대해 Log scaling  
use_index : 객체의 색인을 눈금 이름으로 사용할지 여부  
rot : 눈금 이름 돌리기 (rotating) 0 ~ 360  
xticks, yticks : x, y축으로 사용할 값  
xlim, ylim : X, Y축의 한계  
grid : 축의 그리드를 표현할지 여부

subplots : 각 column에 독립된 subplot 그리기  
sharex, sharey : subplots=True 이면 같은 X,Y축을 공유하고 눈금과 한계를 연결  
figsize : 생성될 그래프의 크기를 tuple로 지정  
title : 그래프의 제목 지정  
legend : subplot의 범례 지정  
sort_columns : column을 알파벳 순서로 그린다.

# Multi Graph 그리기

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(1, 3, 1)
plt.plot(data)
plt.subplot(1, 3, 2)
plt.hist(hist_data, bins = 20)
plt.subplot(1, 3, 3)
plt.scatter(scat_data, np.arange(30) + 3 * np.random.randn(30))
plt.show()
```

---

    - 색상: b(파란색), g(초록색), r(빨간색), c(청록색), y(노란색), k(검은색), w(흰색)
    - 마커: o(원), v(역삼각형), ^(삼각형), s(네모), +(플러스), .(점)

```python
plt.plot(data, "g.-", )
plt.show()
```

```python
plt.plot(data, "r+", )
plt.show()
```

# 그래프 사이즈 조절

- plt.figure 안에 figsize를 이용하여 가로, 세로 길이 조절 가능 (inch 단위)

```python
plt.figure(figsize = (10, 5))  # 사이즈 조절 
plt.plot(data, "r", )
plt.show()
```

- 여러 그래프 그리고 그에 대한 크기 조절

```python
plt.figure(figsize = (13,5))  # 사이즈 조절 (가로, 세로 크기 )

plt.subplot(1, 3, 1)
plt.plot(data)
plt.subplot(1, 3, 2)
plt.hist(hist_data, bins = 20)
plt.subplot(1, 3, 3)
plt.scatter(scat_data, np.arange(30) + 3 * np.random.randn(30))
plt.show()
```

## 그래프 겹치기 + legend 달기

```python
data2 = np.random.randn(30).cumsum()

plt.plot(data2, "k--")
plt.plot(data2, 'k--', drawstyle = 'steps-post')
plt.show()

# 같은데이터를 곂쳐서 볼 수 있다. subplot 안해주면 겹쳐서 나온다. 
```

```python
data2 = np.random.randn(30).cumsum()

plt.plot(data2, "k--", label = 'Default')
plt.plot(data2, 'k', drawstyle = 'steps-post', label = 'steps-post')
plt.legend()  # 무엇을 의미 하는지 표시 
plt.show()

# 같은데이터를 곂쳐서 볼 수 있다. subplot 안해주면 겹쳐서 나온다. 
```

## 이름 달기

```python
plt.plot(np.random.randn(500).cumsum(), 'k')
plt.title("Random Graph")
plt.xlabel('Srages')
plt.ylabel('Values')
plt.show()
```

---
## 종합

```python
plt.figure(figsize = (13,5))

plt.plot(np.random.randn(100).cumsum(), 
        'k', label='one')
plt.plot(np.random.randn(100).cumsum(), 
        'r--', label='two')
plt.plot(np.random.randn(100).cumsum(), 
        'b.', label='three')

plt.legend(loc='best')
plt.show()
```
